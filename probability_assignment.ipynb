### A1 — Website Traffic
**Distribution: Poisson**

Average rate λ = 200 requests/min. Find P(X > 220).

```python
# A1 — Poisson: website traffic
# λ = 200 requests/minute
# P(X > 220) = 1 - P(X <= 220)

lam = 200
p_more_than_220 = 1 - stats.poisson.cdf(220, mu=lam)
print(f'P(X > 220) = {p_more_than_220:.6f}')
print(f'As percentage: {p_more_than_220*100:.4f}%')

# Plot PMF around the mean
x = np.arange(160, 241)
pmf = stats.poisson.pmf(x, mu=lam)

fig, ax = plt.subplots()
ax.bar(x, pmf, color='steelblue', alpha=0.7, width=0.8, label='PMF')
ax.bar(x[x > 220], pmf[x > 220], color='tomato', alpha=0.9, width=0.8, label='P(X > 220)')
ax.axvline(lam, color='black', linestyle='--', linewidth=1, label=f'λ = {lam}')
ax.set_xlabel('Requests per minute')
ax.set_ylabel('Probability')
ax.set_title('A1 — Poisson(λ=200): Website Traffic PMF')
ax.legend()
plt.tight_layout()
plt.savefig('a1_poisson.png', dpi=120)
plt.show()
```

### A2 — Quality Control
**Distribution: Binomial**

n=50 bolts, p=0.02 defective. Find P(X = 3).

```python
# A2 — Binomial: quality control
# n=50 bolts, p=0.02 probability of defect
# P(X = 3 defective)

n, p = 50, 0.02
p_exactly_3 = stats.binom.pmf(3, n=n, p=p)
print(f'P(X = 3) = {p_exactly_3:.6f}')
print(f'As percentage: {p_exactly_3*100:.4f}%')

# Plot PMF
x = np.arange(0, 11)
pmf = stats.binom.pmf(x, n=n, p=p)

fig, ax = plt.subplots()
ax.bar(x, pmf, color='steelblue', alpha=0.7, width=0.5)
ax.bar([3], [stats.binom.pmf(3, n, p)], color='tomato', alpha=0.9, width=0.5, label='P(X=3)')
ax.set_xlabel('Number of defective bolts')
ax.set_ylabel('Probability')
ax.set_title('A2 — Binomial(n=50, p=0.02): Defective Bolts PMF')
ax.legend()
plt.tight_layout()
plt.savefig('a2_binomial.png', dpi=120)
plt.show()
```

### A3 — Delivery Times
**Distribution: Normal**

X ~ N(μ=45, σ=8). Find P(X > 60) and P(40 < X < 50).

```python
# A3 — Normal: delivery times
# X ~ N(mu=45, sigma=8)

mu, sigma = 45, 8
dist = stats.norm(loc=mu, scale=sigma)

p_gt_60 = 1 - dist.cdf(60)
p_40_to_50 = dist.cdf(50) - dist.cdf(40)

print(f'P(X > 60)       = {p_gt_60:.6f}  ({p_gt_60*100:.3f}%)')
print(f'P(40 < X < 50)  = {p_40_to_50:.6f}  ({p_40_to_50*100:.3f}%)')

x = np.linspace(15, 75, 400)
pdf = dist.pdf(x)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: P(X > 60)
axes[0].plot(x, pdf, color='steelblue', linewidth=2)
axes[0].fill_between(x, pdf, where=(x > 60), color='tomato', alpha=0.6, label=f'P(X>60) = {p_gt_60:.4f}')
axes[0].axvline(mu, color='black', linestyle='--', linewidth=1)
axes[0].set_title('P(X > 60 min)')
axes[0].set_xlabel('Delivery time (min)')
axes[0].set_ylabel('Density')
axes[0].legend()

# Right: P(40 < X < 50)
axes[1].plot(x, pdf, color='steelblue', linewidth=2)
axes[1].fill_between(x, pdf, where=((x > 40) & (x < 50)), color='mediumseagreen', alpha=0.6,
                     label=f'P(40<X<50) = {p_40_to_50:.4f}')
axes[1].axvline(mu, color='black', linestyle='--', linewidth=1)
axes[1].set_title('P(40 < X < 50 min)')
axes[1].set_xlabel('Delivery time (min)')
axes[1].legend()

plt.suptitle('A3 — Normal(μ=45, σ=8): Delivery Times', y=1.02)
plt.tight_layout()
plt.savefig('a3_normal.png', dpi=120, bbox_inches='tight')
plt.show()
```

### A4 — Customer Arrivals
**Distribution: Poisson**

Rate = 10/hour = 1/6 per minute. Find P(X=0) in 6 minutes.

```python
# A4 — Poisson: customer arrivals
# Rate = 10 customers/hour
# In 6 minutes: lambda = 10 * (6/60) = 1
# P(X = 0) = no customers in 6 minutes

lam_6min = 10 * (6 / 60)   # = 1.0
p_zero = stats.poisson.pmf(0, mu=lam_6min)

print(f'λ for 6 minutes = {lam_6min}')
print(f'P(X = 0)        = {p_zero:.6f}  ({p_zero*100:.3f}%)')
# Also: e^(-1) = 0.3679 — matches!
print(f'Check: e^(-λ)   = {np.exp(-lam_6min):.6f}')

x = np.arange(0, 8)
pmf = stats.poisson.pmf(x, mu=lam_6min)

fig, ax = plt.subplots()
ax.bar(x, pmf, color='steelblue', alpha=0.7, width=0.5)
ax.bar([0], [p_zero], color='tomato', alpha=0.9, width=0.5, label=f'P(X=0) = {p_zero:.4f}')
ax.set_xlabel('Number of customers in 6 min')
ax.set_ylabel('Probability')
ax.set_title('A4 — Poisson(λ=1): Arrivals in 6 Minutes')
ax.legend()
plt.tight_layout()
plt.savefig('a4_poisson_arrivals.png', dpi=120)
plt.show()
```

### A5 — Class Average Score (CLT)
**Distribution: Normal (by CLT)**

35 students. Assuming population mean μ and std σ, by CLT the sample mean X̄ ~ N(μ, σ²/n).

```python
# A5 — Central Limit Theorem
# 35 students. Assume scores ~ Uniform(0,100) to simulate an unknown distribution.
# CLT says: sample mean X_bar ~ N(mu, sigma^2 / n)

np.random.seed(42)
n = 35
# Simulate: draw 10000 classes of 35 students from a skewed distribution (e.g. Beta)
# to show CLT works even when underlying distribution is NOT normal
pop_dist = stats.beta(2, 5)  # skewed, definitely not normal
pop_scores = pop_dist.rvs(100000) * 100  # scale to 0-100

pop_mean = pop_scores.mean()
pop_std  = pop_scores.std()
se       = pop_std / np.sqrt(n)   # standard error

sample_means = [np.random.choice(pop_scores, size=n, replace=False).mean() for _ in range(5000)]

print(f'Population mean  = {pop_mean:.2f}')
print(f'Population std   = {pop_std:.2f}')
print(f'Standard error   = {se:.2f}  (= sigma / sqrt(n))')
print(f'Simulated mean of sample means = {np.mean(sample_means):.2f}')
print(f'Simulated std  of sample means = {np.std(sample_means):.2f}')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: population distribution (skewed)
axes[0].hist(pop_scores, bins=50, density=True, color='steelblue', alpha=0.7, edgecolor='white')
axes[0].set_title('Population Score Distribution (skewed Beta)')
axes[0].set_xlabel('Score')
axes[0].set_ylabel('Density')

# Right: distribution of sample means → should look normal
axes[1].hist(sample_means, bins=50, density=True, color='mediumseagreen', alpha=0.7, edgecolor='white', label='Sample means')
x_norm = np.linspace(min(sample_means), max(sample_means), 300)
axes[1].plot(x_norm, stats.norm.pdf(x_norm, loc=pop_mean, scale=se),
             color='black', linewidth=2, label=f'N({pop_mean:.1f}, {se:.2f}²)')
axes[1].set_title('Distribution of Sample Means (n=35) — CLT')
axes[1].set_xlabel('Sample mean score')
axes[1].set_ylabel('Density')
axes[1].legend()

plt.suptitle('A5 — Central Limit Theorem: Even a Skewed Population Gives Normal Sample Means', y=1.02)
plt.tight_layout()
plt.savefig('a5_clt.png', dpi=120, bbox_inches='tight')
plt.show()
```

### C1 — Base Rate Fallacy

A medical test is 99% accurate (sensitivity = 99%, specificity = 99%).
Disease affects 1 in 10,000 people.

**Bayes' theorem calculation:**

- P(disease) = 0.0001
- P(positive | disease) = 0.99 (true positive rate)
- P(positive | no disease) = 0.01 (false positive rate)

P(disease | positive) = P(pos|disease) × P(disease) / P(positive)

In 10,000 people: ~1 has the disease, ~99 false positives → only 1 in 100 positives is real.

```python
# C1 — Base rate fallacy with numbers

prevalence   = 1 / 10000   # 1 in 10,000
sensitivity  = 0.99        # P(positive | disease) — true positive rate
specificity  = 0.99        # P(negative | no disease)
fpr          = 1 - specificity  # false positive rate = 0.01

# Bayes theorem
p_pos_given_disease    = sensitivity
p_pos_given_no_disease = fpr
p_disease              = prevalence
p_no_disease           = 1 - prevalence

# P(positive) — total probability of testing positive
p_positive = p_pos_given_disease * p_disease + p_pos_given_no_disease * p_no_disease

# P(disease | positive) — what we actually care about
p_disease_given_positive = (p_pos_given_disease * p_disease) / p_positive

print('=== Base Rate Fallacy ===' )
print(f'Disease prevalence:          {prevalence} (1 in 10,000)')
print(f'Test sensitivity:            {sensitivity} (99%)')
print(f'False positive rate:         {fpr} (1%)')
print()
print(f'P(positive):                 {p_positive:.6f}')
print(f'P(disease | positive):       {p_disease_given_positive:.6f} = {p_disease_given_positive*100:.2f}%')
print()
print('In plain English:')
print(f'  If you test positive, there is only a {p_disease_given_positive*100:.1f}% chance you actually have the disease.')
print(f'  About {(1-p_disease_given_positive)*100:.1f}% of positive results are FALSE POSITIVES.')
print()
# Concrete numbers for 1 million people
n = 1_000_000
true_cases    = int(n * prevalence)
true_pos      = int(true_cases * sensitivity)
false_pos     = int((n - true_cases) * fpr)
print(f'Concrete: In {n:,} people tested:')
print(f'  True positives:  {true_pos}')
print(f'  False positives: {false_pos}')
print(f'  So {true_pos} real cases out of {true_pos+false_pos} positive tests.')
```

### C2 — simulate_clt() Function

```python
# C2 — simulate_clt: works for any scipy.stats distribution

def simulate_clt(distribution, params, n_samples, n_simulations=5000):
    """
    Simulates the Central Limit Theorem for any scipy.stats distribution.

    Args:
        distribution : a scipy.stats distribution class (e.g. stats.expon)
        params       : dict of keyword args for the distribution (e.g. {'scale': 2})
        n_samples    : sample size per simulation (the n in CLT)
        n_simulations: how many samples to draw (default 5000)

    Returns:
        sample_means: array of simulated sample means
    """
    dist = distribution(**params)

    # Draw n_simulations samples each of size n_samples, take the mean
    raw_samples  = dist.rvs(size=(n_simulations, n_samples))
    sample_means = raw_samples.mean(axis=1)

    # Theoretical normal approximation from CLT
    mu_theory = dist.mean()
    se_theory = dist.std() / np.sqrt(n_samples)

    # Normality test on the sample means
    stat, p_val = normaltest(sample_means)

    # Plot
    fig, axes = plt.subplots(1, 2, figsize=(13, 4))

    # Left: underlying distribution shape
    x_pop = np.linspace(dist.ppf(0.001), dist.ppf(0.999), 400)
    try:
        axes[0].plot(x_pop, dist.pdf(x_pop), color='steelblue', linewidth=2)
        axes[0].set_ylabel('PDF')
    except Exception:
        axes[0].hist(dist.rvs(5000), bins=50, density=True, color='steelblue', alpha=0.7)
    axes[0].set_title(f'Underlying Distribution\n{distribution.name}({params})')
    axes[0].set_xlabel('Value')

    # Right: histogram of sample means + normal overlay
    axes[1].hist(sample_means, bins=60, density=True,
                 color='mediumseagreen', alpha=0.7, edgecolor='white',
                 label=f'Sample means (n={n_samples})')
    x_norm = np.linspace(sample_means.min(), sample_means.max(), 400)
    axes[1].plot(x_norm, stats.norm.pdf(x_norm, mu_theory, se_theory),
                 color='black', linewidth=2,
                 label=f'N({mu_theory:.2f}, {se_theory:.3f}²) — CLT')
    axes[1].set_title(f'Sample Means Distribution (n={n_samples})\nnormaltest p={p_val:.4f}')
    axes[1].set_xlabel('Sample mean')
    axes[1].set_ylabel('Density (normalised)')
    axes[1].legend(fontsize=9)

    plt.suptitle(f'CLT Simulation — {distribution.name}, n={n_samples}, {n_simulations} simulations', y=1.02)
    plt.tight_layout()
    fname = f'clt_{distribution.name}_n{n_samples}.png'
    plt.savefig(fname, dpi=120, bbox_inches='tight')
    plt.show()

    print(f'Theoretical μ  = {mu_theory:.4f},  simulated mean = {sample_means.mean():.4f}')
    print(f'Theoretical SE = {se_theory:.4f},  simulated std  = {sample_means.std():.4f}')
    print(f'normaltest p-value = {p_val:.6f}', end=' ')
    print('(normal ✓)' if p_val > 0.05 else '(not yet normal — try larger n)')

    return sample_means


# --- Test 1: Exponential distribution (very skewed) ---
print('=== Test 1: Exponential(scale=2), n=5 (small) ===')
_ = simulate_clt(stats.expon, {'scale': 2}, n_samples=5)

print('\n=== Test 2: Exponential(scale=2), n=40 (large enough for CLT) ===')
_ = simulate_clt(stats.expon, {'scale': 2}, n_samples=40)

print('\n=== Test 3: Uniform distribution, n=30 ===')
_ = simulate_clt(stats.uniform, {}, n_samples=30)
```

### C3 — Exponential Distribution: Why Mean is Misleading

Purchase amounts that follow an Exponential distribution are **heavily right-skewed** — most customers spend a little, a few spend a lot. The mean is pulled up by the extreme high spenders.

Why it's misleading for investors:
- The mean purchase amount could be \$50, but over 60% of customers actually spend less than \$50
- In an Exponential(λ) distribution, mean = 1/λ, but the **median** = ln(2)/λ ≈ 0.693 × mean
- The median is always lower than the mean — so the 'average customer' the investor imagines doesn't exist

Better metrics: median purchase amount, percentile breakdown (p25/p50/p75/p90), and % of revenue from top 10% of spenders.

```python
# C3 — Exponential distribution: mean vs median

scale = 50   # mean purchase = $50
dist  = stats.expon(scale=scale)
samples = dist.rvs(10000, random_state=42)

mean_val   = samples.mean()
median_val = np.median(samples)
p75_val    = np.percentile(samples, 75)
p90_val    = np.percentile(samples, 90)

print(f'Mean purchase:          ${mean_val:.2f}  <- what investors see')
print(f'Median purchase:        ${median_val:.2f}  <- what the typical customer actually spends')
print(f'P75:                    ${p75_val:.2f}')
print(f'P90:                    ${p90_val:.2f}')
print(f'% customers below mean: {(samples < mean_val).mean()*100:.1f}%  <- majority spend less than the mean')

fig, ax = plt.subplots(figsize=(9, 4))
ax.hist(samples, bins=80, density=True, color='steelblue', alpha=0.6, edgecolor='white')
ax.axvline(mean_val,   color='tomato',        linewidth=2, label=f'Mean ${mean_val:.0f}')
ax.axvline(median_val, color='mediumseagreen', linewidth=2, linestyle='--', label=f'Median ${median_val:.0f}')
ax.set_xlabel('Purchase amount ($)')
ax.set_ylabel('Density')
ax.set_title('C3 — Exponential: Mean vs Median for Purchase Amounts')
ax.legend()
ax.set_xlim(0, 300)
plt.tight_layout()
plt.savefig('c3_exponential.png', dpi=120)
plt.show()
```

```python
# Part D — CLT simulation as it would appear in AI output,
# then verified with normaltest

# The AI would explain CLT roughly like this:
# "If you take many samples of the same size from ANY population,
#  the distribution of those sample AVERAGES will look like a bell curve
#  (Normal distribution), regardless of what the original data looks like."

# Why it matters for A/B testing:
# When we run an A/B test, we compute average metrics (avg click rate, avg revenue).
# CLT tells us these averages are approximately Normal, which is exactly what
# lets us use t-tests / z-tests to compute p-values.
# Without CLT, we'd need to know the exact distribution of user behaviour.

# Simulation: verify with scipy.stats.normaltest
np.random.seed(0)

print('=== Verifying CLT with normaltest ===')
print()
print('Underlying distribution: Exponential (very skewed, definitely not normal)')
print()

for n in [5, 15, 30, 50, 100]:
    means = stats.expon(scale=1).rvs(size=(5000, n)).mean(axis=1)
    stat, p = normaltest(means)
    conclusion = 'normal ✓' if p > 0.05 else 'NOT yet normal'
    print(f'  n={n:4d}: normaltest p = {p:.6f}  → {conclusion}')

print()
print('Key CLT-to-p-value connection:')
print('  p-values in t-tests / z-tests assume the test statistic follows a Normal distribution.')
print('  This assumption holds because of CLT — sample means are approximately Normal')
print('  as long as n is large enough (typically n >= 30 for mildly skewed distributions).')
print('  This is WHY we can compute p-values without knowing the exact population distribution.')
```
